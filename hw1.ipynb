```python
#this will help with later calculations and column names won't have to be repeated
cols = ['260: Star Wars: Episode IV - A New Hope (1977)',
       '1210: Star Wars: Episode VI - Return of the Jedi (1983)',
       '356: Forrest Gump (1994)', '318: Shawshank Redemption, The (1994)',
       '593: Silence of the Lambs, The (1991)', '3578: Gladiator (2000)',
       '1: Toy Story (1995)', '2028: Saving Private Ryan (1998)',
       '296: Pulp Fiction (1994)', '1259: Stand by Me (1986)',
       '2396: Shakespeare in Love (1998)', '2916: Total Recall (1990)',
       '780: Independence Day (ID4) (1996)', '541: Blade Runner (1982)',
       '1265: Groundhog Day (1993)', '2571: Matrix, The (1999)',
    "527: Schindler's List (1993)", '2762: Sixth Sense, The (1999)',
       '1198: Raiders of the Lost Ark (1981)', '34: Babe (1995)']
```

```python
#Q1:Mean Rating: Calculate the mean rating for each movie, order with the highest rating listed first, and submit the top three (along with the mean scores for the top two).
df_mean = pd.DataFrame(df[cols].mean(), columns=['mean_values']).sort_values(by='mean_values', ascending=False)
df_mean
```

```python
#Q2: Rating Count (popularity): Count the number of ratings for each movie, order with the most number of ratings first, and submit the top three (along with the counts for the top two). 
df_count = pd.DataFrame(df[cols].count(), columns=['count_ratings']).sort_values(by='count_ratings', ascending=False)
df_count
```

```python
#Q3: % of ratings 4+ (liking): Calculate the percentage of ratings for each movie that are 4 or higher. Order with the highest percentage first, and submit the top three (along with the percentage for the top two).
df_rating = pd.DataFrame(df[cols][df[cols]>=4].count()/df[cols].count(), columns=['ratings']).sort_values(by='ratings', ascending=False)
df_rating
```

```python
#Q4: Top movies for someone who has seen Toy Story
df_toystory = df[df['1: Toy Story (1995)'].notna()]
df_toystory = df_toystory.drop('Gender (1 =F, 0=M)', axis=1)
df_toystory
```

```python
#calculate user count who watched Toy Story
toystory_rating=df_toystory['1: Toy Story (1995)'].count()
toystory_rating
```

```python
#recommendations for someone who watched toy story
toystory_rec = pd.DataFrame(df_toystory[cols].count()/toystory_rating, columns=['ratings']).sort_values(by='ratings', ascending=False)
toystory_rec
```

```python
#Q5: Correlation with Toy Story
df_corr= pd.DataFrame(df[df.columns[2:-1]].apply(lambda x: x.corr(df['1: Toy Story (1995)'])), columns=['correlation']).sort_values(by='correlation', ascending=False)
df_corr
```

```python
#Q6: Mean rating difference by gender:  First, recompute the mean rating for each movie separately for males and for females.  And calculate the overall mean rating (across all ratings) for males and females. Submit the two movies that have the greatest differences 
#calculate Female mean rating
female_mean= pd.DataFrame(df[df['Gender (1 =F, 0=M)']>0][cols].mean(), columns=['F_mean'])
female_mean
```

```python
#calculate male mean rating
male_mean= pd.DataFrame(df[df['Gender (1 =F, 0=M)']==0][cols].mean(), columns=['M_mean'])
male_mean
```

```python
#merge 2 dataframes to calculate their mean difference
df_genders= pd.concat([female_mean, male_mean], axis=1)
df_genders['mean_diff']= df_genders['F_mean']-df_genders['M_mean']
df_genders
```

```python
# difference in overall rating averages (female average - male average).
total_mean_diff= df_genders['F_mean'].sum()-df_genders['M_mean'].sum()
total_mean_diff
```
